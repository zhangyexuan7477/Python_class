### Dictionaries

- Dictionaries are unordered collections of key-value pairs.
- Each key in a dictionary must be unique and immutable (e.g., strings, numbers, tuples).
- Values can be of any data type.
- Dictionaries are defined using curly braces {} with key-value pairs separated by colons :
- Dictionaries are often used as building blocks for more complex data structures (dataframe).

```python
my_dict = {
   "name": "Alice",
   "age": 30,
   "city": "New York"
}
print(my_dict)
```

```python
# accessing Values
my_dict = {
   "name": "Alice",
   "age": 30,
   "city": "New York"
}
print(my_dict["name"])
```

```python
# adding/updating key-value pairs
my_dict = {
   "name": "Alice",
   "age": 30,
   "city": "New York"
}
my_dict["occupation"] = "Engineer"
print(my_dict)
```

```python
# removing key-value pairs
my_dict = {
   "name": "Alice",
   "age": 30,
   "city": "New York"
}
del my_dict["age"]
print(my_dict)
```

```python
# iterating through dictionaries
for key in my_dict:  # Iterating through keys
       print(key)

for value in my_dict.values():  # Iterating through values
       print(value)

for key, value in my_dict.items():  # Iterating through items
       print(key, value)
```

**Lambda function**

Lambda functions are small, anonymous functions defined using the 'lambda' keyword.

Lambda functions are defined in a single line, making the code more compact.

Inline Usage: They can be used directly within functions or expressions.

```python
# Example of a simple lambda function
square = lambda x: x ** 2
print(f"The square of 5 is: {square(5)}")
```

```python
# Example of a lambda function with multiple arguments
add = lambda a, b: a + b
print(f"The sum of 3 and 7 is: {add(3, 7)}")
```

```python
# Example of using lambda functions with built-in functions
# The map function is a built-in Python function. It takes two arguments:
# a function and an iterable (like a list).
# It applies the function to each element
# of the iterable and returns a map object (an iterator).
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(f"The squares of the numbers are: {squared_numbers}")
```

```python
filtered_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(f"Even numbers in the list are: {filtered_numbers}")
```

```python
converter = lambda x : x*2 if x < 10 else (x*3 if x < 20 else x)

print('convert 5 to : ', converter(5))
print('convert 5 to : ', converter(25))
```

```python
# Advanced example of a lambda function
# Using a lambda function to sort a list of tuples by the first element
pairs = [(1, 'one'), (3, 'three'), (2, 'two')]
pairs_sorted = sorted(pairs, key=lambda pair: pair[0])
print(f"Pairs sorted by second element: {pairs_sorted}")
```

**Recursive functions (advanced)**

Recursive functions are functions that call themselves to solve smaller instances of a problem.

They are useful for problems that can be divided into similar sub-problems, such as factorials, Fibonacci sequences,...

Key Points About Recursive Functions:
- Base Case: The condition that stops the recursion.
- Recursive Case: The function calls itself with modified arguments to approach the base case.

```python
# Example of a recursive function: Calculating Factorial
def factorial_recursive(n):
    """Calculate the factorial of a number using recursion."""
    if n == 0:
        return 1  # Base case
    else:
        return n * factorial_recursive(n - 1)  # Recursive case

# Test the recursive function
num = 5
print(f"The factorial of {num} (calculated recursively) is: {factorial_recursive(num)}")
```

```python
# Example of a recursive function: Fibonacci Sequence
def fibonacci_recursive(n):
    """Return the nth Fibonacci number using recursion."""
    if n <= 1:
        return n  # Base case
    else:
        return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

# Test the recursive function
fib_num = 6
print(f"The {fib_num}th Fibonacci number (calculated recursively) is: {fibonacci_recursive(fib_num)}")
```

## Error handling

Error handling is crucial for creating robust programs that can manage unexpected situations gracefully.

Key Points About Error Handling:
- Use `try` to write the code that might raise an exception.
- Use `except` to handle the exception.
- Optionally use `else` for code to execute if no exception occurs.
- Use `finally` to write code that will execute no matter what.

```python
# Simple error handling
try:
  print(j)
except:
  print("An exception occurred")
```

```python
# except: Catches all exceptions
# except ValueError:Catches specific exceptions This handler only catches ValueError exceptions. Other types of exceptions will not be caught.
try:
  x = int(input("Enter a number: "))
except ValueError:
  print("Invalid input. Please enter a valid number.")
```

```python
# Example of more complete error handling
try:
    numerator = int(input("Enter numerator: "))
    denominator = int(input("Enter denominator: "))
    result = numerator / denominator
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
except ValueError:
    print("Error: Invalid input. Please enter integers only.")
else:
    print(f"Result: {result}")
finally:
    print("Execution completed.")
```

```python
# Advanced Example: Custom Exception
class NegativeNumberError(Exception):
    """Custom exception for negative numbers."""
    pass

def check_positive(number):
    if number < 0:
        raise NegativeNumberError("Negative numbers are not allowed.")
    return number

try:
    num = int(input("Enter a positive number: "))
    print(f"You entered: {check_positive(num)}")
except NegativeNumberError as e:
    print(f"Error: {e}")
except ValueError:
    print("Error: Invalid input. Please enter an integer.")
```

```python
class TooLargeError(Exception):
    """Custom exception for input greater than 100."""
    pass

try:
    num = int(input("Enter a number (<= 100): "))
    if num > 100:
        raise TooLargeError("The number is greater than 100.")
    print(f"You entered: {num}")
except TooLargeError as e:
    print(f"Error: {e}")
except ValueError:
    print("Error: Invalid input. Please enter an integer.")
```
